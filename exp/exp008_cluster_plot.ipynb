```python
df["year"] = df["day"].apply(lambda x: x.year)
df_2000 = df[df["year"] == 2000]
df_2001 = df[df["year"] == 2001]
df_2002 = df[df["year"] == 2002]
grouped_times_2000 = df_2000.groupby("name")["seconds_since_midnight"].apply(list).reset_index()
grouped_times_2001 = df_2001.groupby("name")["seconds_since_midnight"].apply(list).reset_index()
grouped_times_2002 = df_2002.groupby("name")["seconds_since_midnight"].apply(list).reset_index()

def create_quantiles(df):
    df["25_quantile"] = df["seconds_since_midnight"].apply(lambda x: np.quantile(np.array(x), .25))
    df["75_quantile"] = df["seconds_since_midnight"].apply(lambda x: np.quantile(np.array(x), .75))
    return df

grouped_times_2000 = create_quantiles(grouped_times_2000)
grouped_times_2001 = create_quantiles(grouped_times_2001)
grouped_times_2002 = create_quantiles(grouped_times_2002)
```
